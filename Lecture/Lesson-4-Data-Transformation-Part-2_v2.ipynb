## 1. mutate(): Creating New Calculated Columns

### ğŸ“– What Is mutate() and Why Does It Exist?

In almost every business dataset, the columns you *have* aren't the columns you *need*. You have `UnitPrice` and `Quantity`, but your boss wants `TotalRevenue`. You have `Amount`, but accounting needs `AmountWithTax`. You have raw numbers, but management wants categories like "High Value" and "Low Value."

**This is what `mutate()` solves.** It takes your existing columns and uses them to create new ones â€” just like writing a formula in a new column in Excel.

---

### ğŸ’¡ The Excel Connection

If you've ever done this in Excel, you've already done what `mutate()` does:

```
Excel:  In cell D2, type =B2*C2, then drag down to fill the whole column
R:      mutate(Total = UnitPrice * Quantity)
```

The difference is that R's `mutate()` applies to **every row automatically** â€” no dragging needed, no chance of missing a row, and no risk of accidentally breaking the formula.

---

### ğŸ¯ The Four Types of Mutations You'll Use in Business

| Type | What It Does | Business Example | Code Pattern |
|------|-------------|-----------------|---------------|
| **Math** | Arithmetic on columns | Total = Price Ã— Qty | `mutate(Total = Price * Qty)` |
| **TRUE/FALSE flag** | Marks rows that meet a condition | Is this order over $500? | `mutate(BigOrder = Amount > 500)` |
| **Two-way label** | Assigns one of two labels | "Big" or "Small" | `mutate(Size = if_else(Amount > 500, "Big", "Small"))` |
| **Multi-way label** | Assigns one of several labels | "High" / "Medium" / "Low" | `mutate(Tier = case_when(...))` |

We'll learn each one step by step with examples.

---

### ğŸ”§ mutate() Syntax Breakdown

```r
new_data <- original_data %>% mutate(NewColumn = some_calculation)
```

| Part | What It Does | Analogy |
|------|-------------|--------|
| `new_data` | Name for your result (you choose this) | The new version of your spreadsheet |
| `<-` | Assignment operator ("save this as...") | Like "Save As" in Excel |
| `original_data` | Your starting dataset | Your original spreadsheet |
| `%>%` | Pipe operator ("and then...") | Pass the data to the next step |
| `mutate()` | The function that creates new columns | "Add a formula column" |
| `NewColumn = calculation` | The name and formula | Column header = formula |

---

### âš ï¸ Important: mutate() Does NOT Change Your Original Data

This is a key concept that trips up beginners. When you run `mutate()`, your original dataset is **untouched**. R creates a *new copy* with the added column. This is actually a safety feature â€” you can never accidentally destroy your raw data.

| What Happens | Code |
|-------------|------|
| Original data stays the same | `customer_transactions` is unchanged |
| New version has the extra column | `new_data` has everything plus the new column |
| To "save" changes permanently, overwrite | `my_data <- my_data %>% mutate(...)` |

---

Let's start with the simplest and most common use of `mutate()`:

### Example 1: Simple Math â€” Calculate Total Revenue Per Order

**Business question**: *"What is the total dollar amount for each transaction?"*

We have `UnitPrice` (price per item) and `Quantity` (number of items). To get the total, we multiply them. This is probably the single most common business calculation you'll ever do.

| Before mutate() | | After mutate() |
|:-:|:-:|:-:|
| UnitPrice = 899.99, Quantity = 1 | â†’ | TotalRevenue = 899.99 |
| UnitPrice = 14.99, Quantity = 2 | â†’ | TotalRevenue = 29.98 |
| UnitPrice = 24.99, Quantity = 3 | â†’ | TotalRevenue = 74.97 |

**What R does behind the scenes**: It goes row by row, multiplies UnitPrice Ã— Quantity, and puts the result in a new column called TotalRevenue.

````r
# Example 1: Calculate total revenue for each transaction
transactions_with_total <- customer_transactions %>%   # Start with our data, AND THEN...
  mutate(TotalRevenue = UnitPrice * Quantity)           # Create new column: Price x Quantity

# Show just the relevant columns so it's easy to see what happened
print("Transactions with new TotalRevenue column:")
transactions_with_total %>%
  select(TransactionID, Product, UnitPrice, Quantity, TotalRevenue) %>%
  print()
````

### ğŸ“– What Just Happened?

Let's trace through what R did:

1. R took the `customer_transactions` data frame (20 rows, 10 columns)
2. For **each row**, it multiplied `UnitPrice Ã— Quantity`
3. It stored the result in a **new column** called `TotalRevenue`
4. We saved this enhanced version as `transactions_with_total` (now has 11 columns)
5. The original `customer_transactions` is still unchanged (still 10 columns)

**Notice**: For rows where `Quantity = 1`, TotalRevenue equals UnitPrice. For rows where `Quantity > 1`, TotalRevenue is larger. That's exactly what we'd expect!

---

### Example 2: Calculating Sales Tax â€” Multiple New Columns at Once

**Business question**: *"Add an 8.25% Texas sales tax column and show the after-tax total."*

This is the kind of calculation you'd do constantly in real business â€” taxes, discounts, commissions, fees, markups. The pattern is always: take an existing number and multiply by a percentage.

**ğŸ’¡ Key concept**: You can create **multiple columns** in one `mutate()` call â€” and later columns can reference earlier ones! R processes them top to bottom.

| What We're Calculating | Formula | Code |
|-----------------------|---------|------|
| Total before tax | UnitPrice Ã— Quantity | `TotalRevenue = UnitPrice * Quantity` |
| Sales Tax (8.25%) | Total Ã— 0.0825 | `SalesTax = TotalRevenue * 0.0825` |
| After-Tax Total | Total + Tax | `AfterTax = TotalRevenue + SalesTax` |

````r
# Example 2: Calculate sales tax and after-tax totals
transactions_with_tax <- customer_transactions %>%     # Start with our data
  mutate(
    TotalRevenue = UnitPrice * Quantity,                 # First: calculate subtotal
    SalesTax = round(TotalRevenue * 0.0825, 2),         # Second: 8.25% tax, rounded to cents
    AfterTaxTotal = round(TotalRevenue + SalesTax, 2)   # Third: add tax to total
  )                                                     # Notice: SalesTax uses TotalRevenue!
                                                        # This works because TotalRevenue was
                                                        # created first (on the line above)

print("Transactions with Sales Tax (first 8 rows):")
transactions_with_tax %>%
  select(TransactionID, Product, TotalRevenue, SalesTax, AfterTaxTotal) %>%
  head(8) %>%                                           # Show first 8 rows to save space
  print()
````

### ğŸ“– What Just Happened?

We created **three new columns** in one `mutate()` call. The order matters:

| Column Created | Depends On | Formula |
|---------------|------------|--------|
| `TotalRevenue` | UnitPrice, Quantity (original columns) | `UnitPrice * Quantity` |
| `SalesTax` | TotalRevenue (just created above!) | `TotalRevenue * 0.0825` |
| `AfterTaxTotal` | TotalRevenue + SalesTax (both new!) | `TotalRevenue + SalesTax` |

**Key insight**: Inside `mutate()`, columns are created **top to bottom**. `SalesTax` can use `TotalRevenue` because `TotalRevenue` was defined on the line above it. If you reversed the order, you'd get an error.

We used `round(value, 2)` to round to 2 decimal places â€” standard practice for dollar amounts.

### Example 3: TRUE/FALSE Flags â€” Marking Orders for Review

**Business question**: *"Which transactions are over $100? Flag them for the sales manager."*

A **flag** is a column that stores `TRUE` or `FALSE` for each row. It's like putting a sticky note on certain invoices that need special attention. Businesses use flags constantly:

| Business Use | Flag Logic |
|-------------|------------|
| Orders needing manager approval | Amount > $500 |
| Customers at risk of churning | Days since last purchase > 90 |
| Products running low on stock | Inventory < 10 |
| Late payments needing follow-up | Days overdue > 30 |

**How it works**: R evaluates the condition for each row and stores `TRUE` or `FALSE`.

| UnitPrice | `UnitPrice > 100` result |
|-----------|-------------------------|
| 899.99 | TRUE âœ… |
| 14.99 | FALSE âŒ |
| 349.99 | TRUE âœ… |

````r
# Example 3: Create a TRUE/FALSE flag for high-value orders
transactions_flagged <- customer_transactions %>%      # Start with our data
  mutate(
    TotalRevenue = UnitPrice * Quantity,                 # Calculate total first
    IsHighValue = TotalRevenue > 100                     # TRUE if over $100, FALSE otherwise
  )

print("Transactions with High-Value Flag:")
transactions_flagged %>%
  select(TransactionID, Product, TotalRevenue, IsHighValue) %>%
  print()

# Bonus: Count how many are high value vs. not
# sum() on TRUE/FALSE counts the TRUEs (TRUE = 1, FALSE = 0)
cat("\nHigh-value transactions (> $100):", sum(transactions_flagged$IsHighValue),
    "out of", nrow(transactions_flagged), "total\n")
````

### Example 4: Two-Way Labels with if_else()

**Business question**: *"Label each order as 'Premium' or 'Standard' based on whether the total is over $100."*

TRUE/FALSE flags are useful for calculations, but they aren't great for reports. Your manager doesn't want to see `TRUE` â€” they want to see `"Premium"`. The `if_else()` function lets you assign **text labels** instead.

---

### ğŸ”§ if_else() Syntax Breakdown

```r
if_else(condition, value_if_TRUE, value_if_FALSE)
```

| Part | What It Means | Example |
|------|-------------|--------|
| `condition` | The test to apply to each row | `TotalRevenue > 100` |
| `value_if_TRUE` | Label when condition IS met | `"Premium"` |
| `value_if_FALSE` | Label when condition is NOT met | `"Standard"` |

**ğŸ’¡ Read it like English**: "IF the total is over 100, THEN call it Premium, ELSE call it Standard."

````r
# Example 4: Two-way labels with if_else()
transactions_labeled <- customer_transactions %>%      # Start with our data
  mutate(
    TotalRevenue = UnitPrice * Quantity,                 # Calculate total
    OrderTier = if_else(TotalRevenue > 100,              # IF total > $100
                        "Premium",                       #   THEN label "Premium"
                        "Standard")                      #   ELSE label "Standard"
  )

print("Transactions with Order Tier Labels:")
transactions_labeled %>%
  select(TransactionID, Product, TotalRevenue, OrderTier) %>%
  print()
````

### Example 5: Multi-Way Labels with case_when()

**Business question**: *"Classify each order into 'High', 'Medium', or 'Low' value tiers."*

`if_else()` only gives you two options. But what if you need three, four, or more categories? That's where `case_when()` comes in. It works like a series of IF-ELSE statements that R checks **from top to bottom**.

---

### ğŸ”§ case_when() Syntax Breakdown

```r
case_when(
  condition1 ~ "Label 1",    # Check this first
  condition2 ~ "Label 2",    # If first didn't match, check this
  TRUE       ~ "Default"     # Everything else gets this label
)
```

| Part | What It Means |
|------|-------------|  
| `condition1` | The first test (e.g., `Amount > 200`) |
| `~` | The tilde means "then assign this label" (think of it as an arrow â†’) |
| `"Label 1"` | The text label to assign (always in quotes) |
| `,` | Comma separates each condition-label pair |
| `TRUE ~ "Default"` | Catch-all for rows that didn't match above (like "else" in Excel IF) |

---

### âš ï¸ Order Matters in case_when()!

R checks conditions **from top to bottom** and stops at the **first match**. Put your most restrictive condition first:

| Order | Code | A $300 order gets... |
|-------|------|---------------------|
| âœ… Correct | `> 200 ~ "High"` first, then `> 50 ~ "Medium"` | **"High"** (correct!) |
| âŒ Wrong | `> 50 ~ "Medium"` first, then `> 200 ~ "High"` | **"Medium"** (wrong! $300 IS > 50, so it matches too early) |

**Rule of thumb**: Start with the biggest/most restrictive number, work down, and always end with `TRUE ~ "Default"`.

````r
# Example 5: Multi-way labels with case_when()
transactions_tiered <- customer_transactions %>%       # Start with our data
  mutate(
    TotalRevenue = UnitPrice * Quantity,                 # Calculate total
    ValueTier = case_when(
      TotalRevenue > 200 ~ "High",                       # Over $200 = High (check first!)
      TotalRevenue > 50  ~ "Medium",                     # $50.01 to $200 = Medium
      TRUE               ~ "Low"                         # $50 and under = Low (catch-all)
    )
  )

print("Transactions with Value Tiers:")
transactions_tiered %>%
  select(TransactionID, Product, TotalRevenue, ValueTier) %>%
  print()
````

### ğŸ“– Let's Trace Through a Few Rows

To make sure you understand how `case_when()` decides, let's walk through the logic for three different orders:

| Row | TotalRevenue | Is it > 200? | Is it > 50? | Hits TRUE? | Result |
|-----|-------------|-------------|------------|-----------|--------|
| Laptop ($899.99) | 899.99 | âœ… YES â†’ stops here | â€” | â€” | **"High"** |
| Jacket ($65.00) | 65.00 | âŒ no | âœ… YES â†’ stops here | â€” | **"Medium"** |
| Novel ($14.99 Ã— 2) | 29.98 | âŒ no | âŒ no | âœ… catches it | **"Low"** |

R checks from the top down and assigns the **first match**. The `TRUE ~` at the bottom catches everything that didn't match any condition above it.

### Example 6: The Complete mutate() â€” Building an Enhanced Transaction Table

**Business question**: *"Give me a complete enhanced view of each transaction: total revenue, sales tax, after-tax total, and a value tier."*

This is what a real analyst would build â€” combining all the `mutate()` patterns into one call.

````r
# Example 6: Build a complete enhanced transaction table
enhanced_transactions <- customer_transactions %>%     # Start with our data
  mutate(
    TotalRevenue = UnitPrice * Quantity,                 # Total order amount
    SalesTax = round(TotalRevenue * 0.0825, 2),         # 8.25% Texas sales tax
    AfterTaxTotal = round(TotalRevenue + SalesTax, 2),  # Grand total with tax
    IsHighValue = TotalRevenue > 100,                    # TRUE/FALSE flag
    ValueTier = case_when(                               # Category label
      TotalRevenue > 200 ~ "High",
      TotalRevenue > 50  ~ "Medium",
      TRUE               ~ "Low"
    )
  )

# Show the key columns from our enhanced data
print("Complete Enhanced Transaction Data:")
enhanced_transactions %>%
  select(TransactionID, CustomerName, Product, TotalRevenue,
         SalesTax, AfterTaxTotal, ValueTier) %>%
  print()
````

## 2. summarize(): Getting the Big Picture Numbers

### ğŸ“– What Is summarize() and Why Does It Exist?

Your VP doesn't want to see 20 rows of transactions. They want to hear: *"Total revenue was $X, average order was $Y, and we served Z customers."* That's **one row** of summary numbers â€” and that's exactly what `summarize()` creates.

The `summarize()` function takes your entire dataset and **collapses it into a single row** of aggregate numbers. Think of it as the **totals row at the bottom of a spreadsheet**.

---

### ğŸ’¡ The Excel Connection

| What You'd Do in Excel | What You'd Write in R |
|-----------------------|----------------------|
| `=SUM(B2:B21)` in a cell below the data | `summarize(Total = sum(Amount))` |
| `=AVERAGE(B2:B21)` in another cell | Add `Avg = mean(Amount)` |
| `=COUNTA(A2:A21)` in another cell | Add `Count = n())` |
| Each formula lives in a separate cell | All metrics in one `summarize()` call |

---

### ğŸ“‹ Summary Functions You'll Use Constantly

| Function | What It Calculates | Excel Equivalent | Example |
|----------|-------------------|-----------------|--------|
| `sum()` | Total of all values | `=SUM()` | `sum(Amount)` â†’ total revenue |
| `mean()` | Average (sum Ã· count) | `=AVERAGE()` | `mean(Amount)` â†’ average order |
| `median()` | Middle value when sorted | `=MEDIAN()` | `median(Amount)` â†’ typical order |
| `n()` | Count of rows | `=COUNTA()` | `n()` â†’ number of transactions |
| `n_distinct()` | Count of unique values | `=COUNTUNIQUE()` | `n_distinct(CustomerID)` â†’ unique customers |
| `min()` | Smallest value | `=MIN()` | `min(Amount)` â†’ smallest order |
| `max()` | Largest value | `=MAX()` | `max(Amount)` â†’ biggest order |
| `sd()` | Standard deviation | `=STDEV()` | `sd(Amount)` â†’ spread of orders |

---

### ğŸ“Š mean() vs. median() â€” A Business Analyst Must Know the Difference

This comes up constantly in business reporting:

| Function | How It Works | Best When... | Example |
|----------|-------------|-------------|--------|
| `mean()` | Adds everything, divides by count | Data is evenly spread | Average monthly sales |
| `median()` | Finds the middle value | Data has outliers/extremes | Typical home price, typical salary |

**Why it matters**: If 9 employees earn $50K and 1 CEO earns $5M:
- `mean()` salary = $545K â€” misleading, nobody actually earns this!
- `median()` salary = $50K â€” accurate for a "typical" employee

**Best practice**: Report both. When they differ a lot, that tells your audience there are outliers pulling the average up or down.

---

### ğŸ”§ summarize() Syntax Breakdown

```r
result <- my_data %>% summarize(MetricName = function(column))
```

| Part | What It Does | Analogy |
|------|-------------|--------|
| `result` | Name for your output | Your one-row executive report |
| `my_data %>%` | Start with your full dataset | All 20 rows of transactions |
| `summarize()` | Collapse all rows into one summary row | Like pressing "AutoSum" in Excel |
| `MetricName = ` | Name you give each metric | The column header in your report |
| `function(column)` | The math to do | Which summary function to use |

**ğŸ’¡ Remember**: `summarize()` without `group_by()` always returns **one single row** â€” the summary of your entire dataset.


## 3. group_by() + summarize(): The Pivot Table of R

### ğŸ“– What Is group_by() and Why Is It So Important?

The `summarize()` above gave us **one row** for the entire dataset. But what if you want summaries for **each region**, or **each product category**, or **each customer** â€” all at once?

That's what `group_by()` does. **It is R's version of an Excel Pivot Table.** It splits your data into groups, then `summarize()` calculates stats for each group separately.

This is arguably **the single most important skill in business analytics**. Almost every business report involves breaking down numbers by category.

---

### ğŸ’¡ The Excel Pivot Table Connection

| In Excel Pivot Table... | In R... |
|------------------------|--------|
| Drag "Region" to the **Rows** area | `group_by(Region)` |
| Drag "Amount" to **Values**, set to SUM | `summarize(Total = sum(Amount))` |
| Drag "Amount" to Values again, set to AVERAGE | Add `Avg = mean(Amount)` |
| **Result**: one row per Region with calculated metrics | Exactly the same! |

### âš ï¸ Always Use `.groups = 'drop'`

After `group_by() + summarize()`, R keeps your data "grouped" behind the scenes. This can cause unexpected behavior if you do more operations later. Always add `.groups = 'drop'` to clean up.

| Code | What Happens |
|------|-------------|
| `summarize(..., .groups = 'drop')` | âœ… Clean result, safe for further analysis |
| `summarize(...)` without it | âš ï¸ Data stays grouped â€” later operations may behave strangely |

**Rule of thumb**: Always include `.groups = 'drop'`. Make it a habit.

## 4. count(): Quick Frequency Counts

### ğŸ“– What Is count()?

`count()` is a shortcut that answers: **"How many of each?"** It groups your data and counts rows per group.

You *could* do this with `group_by() + summarize(n = n())`, but `count()` is cleaner for simple counting.

---

### ğŸ’¡ count() vs. group_by() + summarize()

| Situation | Use This | Why |
|-----------|----------|-----|
| You ONLY need a count | `count()` | Less typing, cleaner |
| You need count AND other metrics | `group_by() + summarize()` | `count()` can only count |

These two produce the **exact same result**:
```r
# Short way:
data %>% count(Region, sort = TRUE)

# Long way (same result):
data %>% group_by(Region) %>% summarize(n = n(), .groups = 'drop') %>% arrange(desc(n))
```

---

### ğŸ”§ count() Syntax

```r
my_data %>% count(column_name, sort = TRUE)
```

| Part | What It Does |
|------|-------------|
| `count()` | Groups by the column and counts rows |
| `column_name` | Which column to group by |
| `sort = TRUE` | Optional: highest count first (always include this) |
| Output column `n` | The count for each group (automatic name) |

````r
# Example 1: Count orders per product category
print("Orders per Product Category:")
customer_transactions %>%
  count(ProductCategory, sort = TRUE) %>%
  print()

# Example 2: Count orders per region
cat("\n")
print("Orders per Region:")
customer_transactions %>%
  count(Region, sort = TRUE) %>%
  print()

# Example 3: Count with percentages (the business-ready version)
cat("\n")
print("Payment Method Breakdown with Percentages:")
customer_transactions %>%
  count(PaymentMethod, sort = TRUE) %>%
  mutate(Pct = round((n / sum(n)) * 100, 1)) %>%
  print()

# Example 4: Cross-tab - count by two columns
cat("\n")
print("Orders by Region and Product Category:")
customer_transactions %>%
  count(Region, ProductCategory, sort = TRUE) %>%
  print()
````

### ğŸ“– Understanding the Percentage Pattern

The formula `round((n / sum(n)) * 100, 1)` will come up over and over. Let's break it down:

| Step | What It Does | Example (if Credit Card has n=8, total=20) |
|------|-------------|---------------------------------------------|
| `n` | The count for this row | 8 |
| `sum(n)` | Total count across ALL rows | 20 |
| `n / sum(n)` | Proportion (decimal) | 8/20 = 0.40 |
| `* 100` | Convert to percentage | 0.40 Ã— 100 = 40.0 |
| `round(..., 1)` | Round to 1 decimal place | 40.0% |

This same pattern works for revenue share, market share, or any "percent of total" calculation. Memorize it!

## 6. Putting It All Together: Business Dashboard

````r
# ================================================================
#                 BUSINESS INTELLIGENCE DASHBOARD
# ================================================================

cat("\n============================================================\n")
cat("          MONTHLY BUSINESS INTELLIGENCE DASHBOARD\n")
cat("          March 2024 | Customer Transaction Analysis\n")
cat("============================================================\n\n")

# --- SECTION 1: Overall Metrics ---
cat("OVERALL BUSINESS METRICS\n")
cat("------------------------------------------------------------\n")

overall <- customer_transactions %>%
  mutate(TotalRevenue = UnitPrice * Quantity) %>%
  summarize(
    total_revenue = round(sum(TotalRevenue), 2),
    total_orders = n(),
    unique_customers = n_distinct(CustomerID),
    avg_order = round(mean(TotalRevenue), 2),
    median_order = round(median(TotalRevenue), 2),
    largest_order = max(TotalRevenue),
    smallest_order = min(TotalRevenue)
  )

cat("  Total Revenue:         $", format(overall$total_revenue, big.mark=","), "\n")
cat("  Total Orders:          ", overall$total_orders, "\n")
cat("  Unique Customers:      ", overall$unique_customers, "\n")
cat("  Average Order Value:   $", overall$avg_order, "\n")
cat("  Median Order Value:    $", overall$median_order, "\n")
cat("  Largest Order:         $", format(overall$largest_order, big.mark=","), "\n")
cat("  Smallest Order:        $", overall$smallest_order, "\n\n")

# --- SECTION 2: Top Product Categories ---
cat("TOP PRODUCT CATEGORIES (by Revenue)\n")
cat("------------------------------------------------------------\n")

top_products <- customer_transactions %>%
  mutate(TotalRevenue = UnitPrice * Quantity) %>%
  group_by(ProductCategory) %>%
  summarize(revenue = round(sum(TotalRevenue), 2), orders = n(), .groups = 'drop') %>%
  mutate(share = round((revenue / sum(revenue)) * 100, 1)) %>%
  arrange(desc(revenue))

for(i in 1:nrow(top_products)) {
  cat("  ", i, ". ", top_products$ProductCategory[i],
      ": $", format(top_products$revenue[i], big.mark=","),
      " (", top_products$orders[i], " orders, ",
      top_products$share[i], "% of revenue)\n", sep="")
}

# --- SECTION 3: Regional Performance ---
cat("\nREGIONAL PERFORMANCE\n")
cat("------------------------------------------------------------\n")

regions <- customer_transactions %>%
  mutate(TotalRevenue = UnitPrice * Quantity) %>%
  group_by(Region) %>%
  summarize(revenue = round(sum(TotalRevenue), 2),
            customers = n_distinct(CustomerID),
            avg_order = round(mean(TotalRevenue), 2),
            .groups = 'drop') %>%
  mutate(share = round((revenue / sum(revenue)) * 100, 1)) %>%
  arrange(desc(revenue))

for(i in 1:nrow(regions)) {
  cat("  ", regions$Region[i],
      ": $", format(regions$revenue[i], big.mark=","),
      " | ", regions$customers[i], " customers",
      " | Avg: $", regions$avg_order[i],
      " | ", regions$share[i], "%\n", sep="")
}

# --- SECTION 4: Customer Value Tiers ---
cat("\nCUSTOMER VALUE TIERS\n")
cat("------------------------------------------------------------\n")

tiers <- customer_transactions %>%
  mutate(TotalRevenue = UnitPrice * Quantity) %>%
  group_by(CustomerID, CustomerName) %>%
  summarize(Total_Spent = sum(TotalRevenue), .groups = 'drop') %>%
  mutate(Tier = case_when(
    Total_Spent >= 500 ~ "Gold",
    Total_Spent >= 200 ~ "Silver",
    TRUE ~ "Bronze"
  )) %>%
  count(Tier) %>%
  mutate(pct = round((n / sum(n)) * 100, 1))

for(i in 1:nrow(tiers)) {
  cat("  ", tiers$Tier[i], ": ", tiers$n[i],
      " customers (", tiers$pct[i], "%)\n", sep="")
}

# --- SECTION 5: Payment Methods ---
cat("\nPAYMENT METHOD BREAKDOWN\n")
cat("------------------------------------------------------------\n")

payments <- customer_transactions %>%
  count(PaymentMethod, sort = TRUE) %>%
  mutate(pct = round((n / sum(n)) * 100, 1))

for(i in 1:nrow(payments)) {
  cat("  ", payments$PaymentMethod[i], ": ",
      payments$n[i], " orders (", payments$pct[i], "%)\n", sep="")
}

cat("\n============================================================\n")
cat("                    END OF DASHBOARD\n")
cat("============================================================\n")
````

## Common Mistakes and How to Fix Them

Everyone makes these mistakes when learning. Knowing them in advance saves frustration.

---

### âš ï¸ Top Mistakes to Avoid

| # | Mistake | What Goes Wrong | How to Fix It |
|---|---------|----------------|---------------|
| 1 | Forgetting `.groups = 'drop'` | Data stays grouped, weird results later | Always add `.groups = 'drop'` in `summarize()` |
| 2 | Using `=` instead of `==` in filter | `filter(x = 5)` causes an error | Use `filter(x == 5)` (double equals) |
| 3 | Forgetting `~` in case_when | `case_when(Amount > 100, "High")` fails | Use `case_when(Amount > 100 ~ "High")` |
| 4 | Missing `TRUE ~` in case_when | Unmatched rows become `NA` (blank) | Always end with `TRUE ~ "Default"` |
| 5 | Wrong order in case_when | Smaller condition catches big values | Put biggest/most restrictive condition first |
| 6 | Forgetting commas between columns in mutate | Syntax error | Separate each new column with a comma |
| 7 | Using a column before creating it | "Column not found" error | Inside `mutate()`, columns are created top-to-bottom |
| 8 | Forgetting `round()` | Output shows 12 decimal places | `round(value, 2)` for dollars, `round(value, 1)` for percentages |

---

### ğŸ”§ case_when() Checklist

| âœ… Check | Correct | âŒ Wrong |
|----------|---------|----------|
| Uses `~` between condition and label | `Amount > 100 ~ "High"` | `Amount > 100, "High"` |
| Pairs separated by commas | `> 100 ~ "High", > 50 ~ "Med"` | Missing comma between pairs |
| `TRUE ~` as last catch-all | `TRUE ~ "Low"` | No `TRUE ~` â†’ unmatched = `NA` |
| Labels in quotes | `~ "High"` | `~ High` (no quotes = error) |
| Biggest condition first | `> 200` first, then `> 100` | `> 100` first catches everything |

---

### ğŸ”§ summarize() Checklist

| âœ… Check | Why |
|----------|-----|
| Used `.groups = 'drop'` | Prevents hidden grouping causing problems |
| Used `round()` on means and percentages | Clean output for reports |
| Gave descriptive metric names | `Total_Revenue` beats `x1` |
| Used `n_distinct()` for unique counts | `n()` counts rows; `n_distinct()` counts unique values |
